```python
batsmen['NO'] = batsmen['NO'].astype(int)
batsmen['Runs'] = batsmen['Runs'].astype(int)
batsmen['Ave'] = batsmen['Ave'].astype(float)
batsmen['SR'] = batsmen['SR'].astype(float)
batsmen['100'] = batsmen['100'].astype(int)
batsmen['50'] = batsmen['50'].astype(int)
batsmen['4s'] = batsmen['4s'].astype(int)
batsmen['6s'] = batsmen['6s'].astype(int)
```

```python
bowlers = pd.read_csv("C:/Users/hp/Downloads/Studies/Data Mining/cp3/Bowling 2023.csv")
bowlers
```

```python
bowlers.drop('BBI', axis=1, inplace=True)

bowlers['Team'] = bowlers['Player'].str.extract(r'\((\w+)\)')
bowlers['Player'] = bowlers['Player'].str.extract(r'([^\(\)]+)')

bowlers = bowlers.replace('-', np.nan)
bowlers.fillna(0, inplace = True)
bowlers.head()
```

```python
bowlers.info()
```

```python
bowlers['Mdns'] = bowlers['Mdns'].astype(int)
bowlers['Runs'] = bowlers['Runs'].astype(int)
bowlers['Wkts'] = bowlers['Wkts'].astype(int)
bowlers['Ave'] = bowlers['Ave'].astype(float)
bowlers['Econ'] = bowlers['Econ'].astype(float)
bowlers['4'] = bowlers['4'].astype(int)
bowlers['5'] = bowlers['5'].astype(int)
```

```python
bowlers['points'] = bowlers['Wkts']*25 + bowlers['4']*8 + bowlers['5']*16 + bowlers['Mdns']*12
bowlers
```

```python
def playingXI(team, batsmen=batsmen, bowlers=bowlers):
    batsmen = batsmen[batsmen['Team']==team]
    bowlers = bowlers[bowlers['Team']==team]

    batsmen = batsmen.sort_values(by='points', ascending=False)
    
    bowlers = bowlers.sort_values(by='points', ascending=False)
    
    merged_data = pd.merge(batsmen[['Player', 'points']], bowlers[['Player', 'points']], on='Player', how='outer')
    merged_data = merged_data.fillna(0)

    merged_data['Final_points'] = np.maximum(merged_data['points_x'], merged_data['points_y'])
    
    merged_data.drop(['points_x','points_y'], axis=1, inplace = True)

    merged_data = merged_data.sort_values(by='Final_points', ascending=False)
    p11 = merged_data.head(11).reset_index().drop('index',axis=1)
    
    return p11['Player']
```

```python
playingXI('IND')
```
